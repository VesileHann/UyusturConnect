````python
#[5]
def draw_neural_net(ax, left, right, bottom, top, layer_sizes):
    n_layers = len(layer_sizes)
    v_spacing = (top - bottom)/float(max(layer_sizes))
    h_spacing = (right - left)/float(len(layer_sizes) - 1)
    # Nodes
    for n, layer_size in enumerate(layer_sizes):
        layer_top = v_spacing*(layer_size - 1)/2. + (top + bottom)/2.
        for m in range(layer_size):
            circle = plt.Circle((n*h_spacing + left, layer_top - m*v_spacing), v_spacing/4.,
                                color='w', ec='k', zorder=4)
            ax.add_artist(circle)
    # Edges
    for n, (layer_size_a, layer_size_b) in enumerate(zip(layer_sizes[:-1], layer_sizes[1:])):
        layer_top_a = v_spacing*(layer_size_a - 1)/2. + (top + bottom)/2.
        layer_top_b = v_spacing*(layer_size_b - 1)/2. + (top + bottom)/2.
        for m in range(layer_size_a):
            for o in range(layer_size_b):
                line = plt.Line2D([n*h_spacing + left, (n + 1)*h_spacing + left],
                                  [layer_top_a - m*v_spacing, layer_top_b - o*v_spacing], c='k')
                ax.add_artist(line)
````

### Backpropagation


**Backpropagation (Geri Yayılım)**, yapay sinir ağlarının eğitilmesinde kullanılan bir optimizasyon algoritmasıdır. Temel olarak, ağın öğrenme sürecinde hataları geriye doğru ileterek ağı günceller.

1. **İleri Yayılım (Forward Propagation)**: Önce, ağ girdilerden başlayarak ileri doğru hesaplamalar yapar ve çıktı üretir. Bu hesaplamalar, ağırlıklar ve aktivasyon fonksiyonları aracılığıyla gerçekleşir.

2. **Hata Hesaplama (Error Calculation)**: Ardından, ağın çıktıları ile gerçek değerler arasındaki hata hesaplanır. Bu hata genellikle bir kayıp fonksiyonu kullanılarak ölçülür.

3. **Geri Yayılım (Backpropagation)**: Hatanın ağ içindeki katmanlara nasıl geri yayılacağı belirlenir. Geri yayılma, zincir kuralı kullanılarak hata gradientleri hesaplanarak yapılır. Bu gradientler, her bir ağırlığın hata üzerindeki etkisini gösterir.

4. **Ağı Güncelleme (Weight Update)**: Elde edilen hata gradientleri kullanılarak ağın ağırlıkları güncellenir. Genellikle bir optimizasyon algoritması (örneğin, Gradient Descent) kullanılarak ağırlıkların güncellenmesi gerçekleştirilir.

---

Yapay sinir ağları için başlangıç ağırlıklarını ve biasları oluşturulur.Verilen kodu incelersek:

1. `num_inputs = len(X_train[0])`: Bu satır, giriş verilerinizdeki özelliklerin sayısını (`X_train` içindeki bir örnekten alınarak) `num_inputs` değişkenine atar. Bu, giriş katmanının nöron sayısını belirler.

2. `hidden_layer_neurons = 13`: Gizli katmandaki nöron sayısını `hidden_layer_neurons` değişkenine atar. Bu, yapay sinir ağınızın içerdiği gizli katman nöronlarının sayısını belirler.

3. `np.random.seed(4)`: Rastgele sayı üretimi için kullanılan rastgele sayı üreteci için bir tohum değeri belirler. Bu, her çalıştırmada aynı rastgele değerleri almanızı sağlar.

4. `b1 = np.random.uniform(-1, 1, hidden_layer_neurons)`: `hidden_layer_neurons` sayısı kadar rastgele sayı üretir ve bu değerleri -1 ile 1 arasında uniform (düzgün) bir dağılıma göre bias (`b1`) vektörü olarak atar. Bias vektörü, her bir gizli katman nöronunun aktivasyon fonksiyonuna eklenen sabit değerlerdir.

5. `w1 = np.random.uniform(-1, 1, (num_inputs, hidden_layer_neurons))`: `num_inputs` ve `hidden_layer_neurons` boyutlarında bir matris oluşturur. Bu matris, giriş özelliklerinden gizli katmana giden ağırlıkları içerir. Ağırlık matrisindeki her bir öğe, bir giriş özelliğinden bir gizli katman nöronuna giden bağı gösterir.

6. `print("Bias (b1):", b1)`: Oluşturulan bias vektörünü ve ağırlık matrisini ekrana yazdırır. Bu, oluşturulan başlangıç değerlerini görmek için kullanılır.

````python
num_inputs = len(X_train[0])  # Giriş özelliklerinin sayısı
hidden_layer_neurons = 13  # Gizli katmandaki nöron sayısı

# Rastgele başlangıç değerleri oluşturma
np.random.seed(4)
b1 = np.random.uniform(-1, 1, hidden_layer_neurons)  # Bias vektörü
w1 = np.random.uniform(-1, 1, (num_inputs, hidden_layer_neurons))  # Ağırlık matrisi

# Oluşturulan ağırlıkların ve bias'ların değerlerini gösterme
print("Bias (b1):", b1)
print("Weight (w1):", w1)
````

````python
num_outputs = 7
b2 = 2*np.random.random(num_inputs) -1
w2 = 2*np.random.random((hidden_layer_neurons, num_outputs)) - 1
# Oluşturulan ağırlıkların ve bias'ların değerlerini gösterme
print("Bias (b2):", b2)
print("Weight (w2):", w2)
````

Bu kod parçası, sigmoid aktivasyon fonksiyonunu görselleştirir. Sigmoid fonksiyonu, yapay sinir ağlarında yaygın olarak kullanılan bir aktivasyon fonksiyonudur. İşte adım adım açıklaması:
    - `xp = np.linspace(-5, 5, 50)`: `-5` ile `5` arasında 50 eşit aralıklı değer oluşturur ve `xp` değişkenine atar. Bu değerler, x ekseninde sigmoid fonksiyonunu çizeceğimiz noktaları temsil eder.
    - `yp = 1 / (1 + np.exp(-xp))`: Her bir `xp` değeri için sigmoid fonksiyonunu hesaplar ve `yp` değişkenine atar. `np.exp(-xp)`, `xp`'nin negatif üssünü alarak her bir `xp` değeri için e üzeri (-xp) hesaplar.
    - `plt.plot(xp, yp)`: `xp` ve `yp` değerlerini kullanarak sigmoid fonksiyonunun grafiğini çizer.
    - `plt.title("Sigmoid Function")`: Grafiğe bir başlık ekler.
    - `plt.xlabel("Input (x)")`: X eksenine "Input (x)" etiketini ekler.
    - `plt.ylabel("Output (σ(x))")`: Y eksenine "Output (σ(x))" etiketini ekler.
    - `plt.grid(True)`: Grafiğe bir ızgara ekler.
    - `plt.show()`: Grafiği görüntüler.

Bu grafik sigmoid fonksiyonunun xp ve yp değerlerine göre çalıştığğını örsel olarak anlamak için kullanılmıştır. Sigmoid fonksiyonu, küçük negatif giriş değerleri için çıktının 0'a, büyük pozitif giriş değerleri için ise 1'e yaklaştığını gösterir. Bu, sinir ağlarının aktivasyonları normalleştirmesi ve çıktıların belirli bir aralıkta kalmasını sağlaması açısından faydalıdır.

````python
# -5 ile 5 arasında 50 eşit aralıklı nokta oluşturur
xp = np.linspace(-5, 5, 50)

# Her bir x değeri için sigmoid fonksiyonunu hesaplar
yp = 1 / (1 + np.exp(-xp))

# xp ve yp noktalarını kullanarak sigmoid fonksiyonunu çizer
plt.plot(xp, yp)
plt.title("Sigmoid Function")
plt.xlabel("x prediction")
plt.ylabel("y prediciton")
plt.grid(True)
plt.show()
````

Bu kod parçası, basit bir iki katmanlı yapay sinir ağı oluşturur ve eğitir. Eğitim süreci boyunca, ileri yayılım ve geri yayılım adımları gerçekleştirilir. İleri yayılım, ağın tahmin yapmasını sağlar, geri yayılım ise ağırlıkları güncelleyerek hatayı minimize eder. Bu süreç, belirtilen epoch sayısı kadar tekrarlanır ve her epoch sonunda hata değeri kaydedilir ve yazdırılır.

````python
#[3]
error = []
b1=0
b2=0
learning_rate = 0.2 # slowly update the network
for epoch in range(1000):
    #ileri yayılım işlemi
    l1 = 1/(1 + np.exp(-(np.dot(X_train, w1) + b1))) # sigmoid function
    l2 = 1/(1 + np.exp(-(np.dot(l1, w2) +b2 )))
    #hata hesaplama
    er = (abs(y_train - l2)).mean()
    #geri yayılım işlemi
    l2_delta = (y_train - l2)*(l2 * (1-l2))
    l1_delta = l2_delta.dot(w2.T) * (l1 * (1-l1))
    #ağırlik ve bias güncelleme
    w2 += l1.T.dot(l2_delta) * learning_rate
    w1 += X_train.T.dot(l1_delta) * learning_rate
    #hata kaydı ve yazdırma
    error.append(er/(epoch*0.1))
    print('Error:', er)
````

- `classifier.add(Dense(output_dim=16, init='uniform', activation='relu', input_dim=12))`:
  - `Dense`: Tam bağlantılı (fully connected) katmanı ekler.
  - `output_dim=16`: Bu katmanda 16 nöron olduğunu belirtir.
  - `init='uniform'`: Ağırlıkların başlangıç değerlerini uniform dağılımla rastgele ayarlamayı belirtir.
  - `activation='relu'`: Aktivasyon fonksiyonu olarak ReLU (Rectified Linear Unit) kullanılır. ReLU, negatif girdileri 0 yaparken pozitif girdileri olduğu gibi bırakır.
  - `input_dim=12`: Giriş katmanının 12 özellikten (input features) oluştuğunu belirtir.

3. Overfitting'i Önlemek için Dropout Katmanı
- `classifier.add(Dropout(p=0.1))`:
  - `Dropout`: Dropout katmanı ekler, bu katman eğitim sırasında bazı nöronları rastgele devre dışı bırakarak overfitting'i önler.
  - `p=0.1`: Dropout oranını %10 olarak belirler, yani her eğitim adımında nöronların %10'u devre dışı bırakılır.

4. İkinci Gizli Katman
- `classifier.add(Dense(output_dim=16, init='uniform', activation='relu'))`:
  - İlk gizli katmanla aynı yapıdadır (16 nöron, uniform başlangıç, ReLU aktivasyon fonksiyonu), ancak bu sefer giriş boyutu belirtilmez çünkü önceki katmanın çıkış boyutu otomatik olarak giriş boyutu olarak kullanılır.

 5. Overfitting'i Önlemek için İkinci Dropout Katmanı
- `classifier.add(Dropout(p=0.1))`: Yine %10 dropout oranı ile overfitting'i önlemek için kullanılır.
6. Çıkış Katmanı
- `classifier.add(Dense(output_dim=7, init='uniform', activation='sigmoid'))`:
  - Çıkış katmanı, 7 sınıf olduğunu belirtir (örneğin, 7 farklı kategori veya sınıf).
  - `activation='sigmoid'`: Çıkış aktivasyon fonksiyonu olarak sigmoid kullanılır, çünkü sigmoid fonksiyonu çıktı değerlerini 0 ile 1 arasında sınırlar, bu da çok sınıflı sınıflandırma problemleri için uygundur.

Bu model, iki gizli katman ve bir çıkış katmanı içeren basit bir yapay sinir ağı modelidir. Her gizli katman ReLU aktivasyon fonksiyonu kullanırken, çıkış katmanı sigmoid aktivasyon fonksiyonu kullanır. Overfitting'i önlemek için dropout katmanları eklenmiştir. Bu model, giriş özelliklerini (12 adet) kullanarak, 7 sınıflı bir sınıflandırma problemi çözmek üzere yapılandırılmıştır.

````python
# Modelin tanımlanması
classifier = Sequential()

# Giriş katmanı ve ilk gizli katmanın eklenmesi
# ReLU (Rectified Linear Unit) aktivasyon fonksiyonu kullanılıyor
classifier.add(Dense(units=16, kernel_initializer='uniform', activation='relu', input_dim=12))

# Overfitting'i önlemek için dropout katmanının eklenmesi
classifier.add(Dropout(rate=0.1))  # %10 dropout oranı

# İkinci gizli katmanın eklenmesi
classifier.add(Dense(units=16, kernel_initializer='uniform', activation='relu'))  # İkinci gizli katman, 16 nöron

# Overfitting'i önlemek için tekrar dropout katmanının eklenmesi
classifier.add(Dropout(rate=0.1))  # %10 dropout oranı

# Çıkış katmanının eklenmesi
classifier.add(Dense(units=7, kernel_initializer='uniform', activation='sigmoid'))  # Çıkış katmanı, 7 nöron ve sigmoid aktivasyon fonksiyonu ile
````

### Çoklu Sınıflandırma için SVM

Bu kod parçası, bir SVM (Destek Vektör Makinesi) modeli kullanarak bir sınıflandırma problemi için test verileri üzerinde tahminler yapar ve ardından doğruluk (accuracy) skorunu hesaplar. Ayrıca, bu tahminlerin sonuçlarını görselleştirmek için bir karışıklık matrisi (confusion matrix) oluşturur.

- **SVM Modeli Oluşturma ve Eğitme**: Kod, `SVC` sınıfından bir SVM modeli oluşturur. Model, "rbf" (radial basis function) çekirdeği kullanarak ve `C=2` parametresiyle eğitilir. Bu model daha sonra `X_train1` ve `y_train1` verileriyle eğitilir.

- **Tahminler ve Doğruluk Skoru Hesaplama**: Oluşturulan SVM modeli, test verileri (`X_test`) üzerinde tahminler yapar ve bu tahminlerle gerçek etiketler (`y_test1`) arasındaki doğruluk skorunu hesaplar. Doğruluk skoru, doğru tahmin edilen örneklerin tüm örnekler içindeki oranını verir.

- **Karışıklık Matrisi Görselleştirme**: Tahminlerin doğruluğunu daha ayrıntılı bir şekilde analiz etmek için bir karışıklık matrisi oluşturulur. Bu matris, modelin her bir sınıf için doğru ve yanlış tahminlerini gösterir. Heatmap olarak görselleştirilir ve matrisin hücrelerine doğruluk (TP: True Positive), yanlış pozitiflik (FP: False Positive), doğruluk (TN: True Negative) ve yanlış negatiflik (FN: False Negative) değerleri yazılır.

Amacı ise SVM modelinin performansını değerlendirmek ve modelin sınıfları doğru bir şekilde tahmin edip edemediğini anlamak içindir. Doğruluk skoru ve karışıklık matrisi, modelin sınıflandırma yeteneğini ve hangi sınıfların daha doğru veya yanlış tahmin edildiğini görmek için kullanılır.

````python
from sklearn import metrics#[6]

# Karışıklık matrisini görselleştirmek için bir fonksiyon tanımlanır.
def plot_confusion_metrix(y_test, model_test):
    cm = metrics.confusion_matrix(y_test, model_test)  # Karışıklık matrisi hesaplanır.
    plt.figure(1)
    plt.clf()
    plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Wistia)  # Matris görselleştirilir.
    classNames = ['Non-User', 'User']
    plt.title('Confusion Matrix')
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    tick_marks = np.arange(len(classNames))
    plt.xticks(tick_marks, classNames)
    plt.yticks(tick_marks, classNames)
    s = [['TN', 'FP'], ['FN', 'TP']]
    for i in range(2):
        for j in range(2):
            plt.text(j, i, str(s[i][j]) + " = " + str(cm[i][j]))  # Matris hücrelerine değerler yazılır.
    plt.show()

# Destek Vektör Makinesi (SVM) modeli oluşturulur ve eğitilir.
from sklearn.svm import SVC
svm = SVC(kernel="rbf", C=2, random_state=0)
svm.fit(X_train1, y_train1.ravel())

# Model ile test verileri üzerinde tahminler yapılır.
pred = svm.predict(X_test)
# Doğruluk skoru hesaplanır.
accu = metrics.accuracy_score(y_test1, pred)
````

   - Bu fonksiyon, bir modelin performansını değerlendirmek için kullanılır. Performans değerlendirmesi genellikle sınıflandırma modelleri için yapılır.
   - İlk olarak, modelin test verileri üzerindeki tahminlerini alır. Bu tahminler, modelin test verileri üzerinde hangi sınıfları doğru veya yanlış tahmin ettiğini gösterir.
   - Ardından, `metrics.confusion_matrix` ile karışıklık matrisini hesaplar. Karışıklık matrisi, modelin her bir sınıf için doğru ve yanlış tahminlerini gösteren bir matristir. Bu matris, modelin hangi sınıfları doğru veya yanlış tahmin ettiğini net bir şekilde gösterir.
   - `metrics.classification_report` ile sınıflandırma raporunu ekrana yazdırır. Bu rapor, modelin sınıflandırma performansını detaylı olarak gösterir. Hassasiyet (precision), özgüllük (recall), F1 skoru gibi metriklerle her bir sınıf için performansı değerlendirir.

2. **`roc_curves(model)` Fonksiyonu**:
   - Bu fonksiyon, ROC eğrisini (Receiver Operating Characteristic curve) çizer. ROC eğrisi, bir sınıflandırma modelinin duyarlılığını (true positive rate) ve özgüllüğünü (false positive rate) gösteren bir grafiktir.
   - İlk olarak, modelin test verileri üzerindeki tahminlerini alır.
   - `roc_curve` ile ROC eğrisinin çizim için gerekli verileri hesaplar. Bu veriler, farklı eşik değerlerine göre modelin duyarlılık ve özgüllük değerlerini içerir.
   - Son olarak, matplotlib kütüphanesi kullanılarak ROC eğrisi çizilir ve AUC (Area Under Curve) değeri ile birlikte görselleştirilir. AUC değeri, ROC eğrisinin altındaki alanı temsil eder ve modelin performansını ölçmek için önemli bir metriktir.

3. **`accuracy(model)` Fonksiyonu**:
   - Bu fonksiyon, modelin doğruluk skorunu hesaplar ve ekrana yazdırır. Doğruluk skoru, modelin test verileri üzerinde doğru tahmin ettiği örneklerin oranını gösterir.
   - `metrics.accuracy_score` ile doğruluk skoru hesaplanır ve ekrana yazdırılır.

Bu fonksiyonlar, bir sınıflandırma modelinin performansını detaylı olarak değerlendirmek, modelin sınıfları doğru tahmin edip edemediğini, hangi sınıfların daha iyi veya daha kötü tahmin edildiğini anlamak ve modelin genel performansını ölçmek için kullanılır. Bir sonraki kod bloğunda yazılı olan kod ile veriyi daha iyi bir şekilde öğrenebiliriz.

````python
from sklearn.metrics import roc_curve, auc

def report_performance(model):
    # Modelin test verileri üzerindeki tahminlerini alır.
    model_test = model.predict(X_test)

    # Karışıklık matrisini ve sınıflandırma raporunu ekrana yazdırır.
    print("\n\nKarışıklık Matrisi:")
    print("{0}".format(metrics.confusion_matrix(y_test, model_test)))
    print("\n\nSınıflandırma Raporu: ")
    print(metrics.classification_report(y_test, model_test))

    # Karışıklık matrisini görselleştirmek için fonksiyonu çağırır.
    plot_confusion_metrix(y_test, model_test)

def roc_curves(model):
    # Modelin test verileri üzerindeki tahminlerini alır.
    predictions_test = model.predict(X_test)

    # ROC eğrisini çizer.
    fpr, tpr, thresholds = roc_curve(predictions_test, y_test)
    roc_auc = auc(fpr, tpr)

    # ROC eğrisini görselleştirir.
    plt.figure()
    plt.plot(fpr, tpr, color='darkorange', lw=1, label='ROC eğrisi (alan = %0.2f)' % roc_auc)
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('Yanlış Pozitif Oranı')
    plt.ylabel('Doğru Pozitif Oranı')
    plt.title('Alıcı İşletim Karakteristiği (ROC)')
    plt.legend(loc="lower right")
    plt.show()

def accuracy(model):
    # Modelin test verileri üzerindeki tahminlerini alır.
    pred = model.predict(X_test1)

    # Modelin doğruluk skorunu hesaplar ve ekrana yazdırır.
    accu = metrics.accuracy_score(y_test, pred)
    print("\nModelin Doğruluğu: ", accu, "\n\n")
````

````python
from lazypredict.Supervised import LazyClassifier
from sklearn.model_selection import train_test_split
feature_col_names = ['EDU_18', 'EDU_Some_college', 'EDU_Cert', 'EDU_Uni', 'EDU_Mas', 'EDU_Doc', 'Age25_34', 'Age35_44', 'Age45_54', 'Age55_64', 'Age65_', 'Gender']
predicted_class_names = ['Benzos']

X = df[feature_col_names].values
y = df[predicted_class_names].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)


# Lazy Predict ile model seçimi ve performans değerlendirmesi
clf = LazyClassifier(verbose=0, ignore_warnings=True, custom_metric=None)
models_train,predictions_train = clf.fit(X_train, X_train, y_train, y_train)
models_test,predictions_test = clf.fit(X_train, X_test, y_train, y_test)
````
